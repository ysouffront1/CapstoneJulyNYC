# CDC NHANES 2013-2014 EDA Part 2: Clustering 

## Overview

The purpose of this notebook is to find the best 'k' value for each dataset using the elbow method and silhouette method. Then to further optimize each model, while prioritizing accuracy and significance of each group of clusters by dataset. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA


from warnings import filterwarnings
filterwarnings('ignore')

pd.options.display.max_columns = 100

#import plotly.express as px
#import plotly.offline as pyo
#pyo.init_notebook_mode()
```

## Demographic Dataset 

#### Importing Data

```python
# importing data
demographic_df = pd.read_csv('~/Downloads/capstone/CapstoneJulyNYC/Data/Clean_Data/demographic_df_clean_filtered.csv')
demographic_df
```

```python
demographic_df.shape
```

```python
# setting SEQN as index 
demographic_df = demographic_df.set_index('SEQN')
demographic_df
```

```python
#Scaling Clean Demographic Data
# 1. Import the scaler
from sklearn.preprocessing import StandardScaler 

# 2. make a scaler
my_standard_scaler = StandardScaler() 

# 3. fit the scaler
my_standard_scaler.fit(demographic_df)

# 4. transform the data. Notice we get back a numpy array even if we put in a dataframe
scaled_demographic_df_data = my_standard_scaler.transform(demographic_df)

scaled_demographic_df = pd.DataFrame(columns=demographic_df.columns, data=scaled_demographic_df_data, index = demographic_df.index)
scaled_demographic_df.head()
```

```python
Xdemographic= scaled_demographic_df
scaler = StandardScaler()
Xscaled_demographic = scaler.fit_transform(Xdemographic)
```

#### Finding Best 'K' Value 

##### Elbow Method

```python
# Look over a range of K values
k_ranges = np.arange(1, 20)

# Instantiate an empty list of inertias
inertia_list = []

# Iterate over ks
for k in k_ranges:
    
    # Instantiate
    k_means_model = KMeans(n_clusters=k)

    # Fit
    k_means_model.fit(Xscaled_demographic)
    
    # Get inertia and add to list
    inertia_list.append(k_means_model.inertia_)  
```

```python
# Visualize
plt.figure()
plt.plot(k_ranges, inertia_list, marker="o", color='purple')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Scree Plot for K-means for Demographic Dataset')
plt.xticks(k_ranges)
plt.show();
```

##### Silhouette Score

```python
# Look over a range of K values
k_range = np.arange(2, 20)

# Instantiate an empty list of inertias
silscores_list = []

# Iterate over ks
for k in k_range:
    #print(f'We are currently on k= {k}')
    #kmeans
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_demographic_df)
    score = silhouette_score(scaled_demographic_df,kmeans.labels_)
    silscores_list.append(silhouette_score(scaled_demographic_df, kmeans.labels_))
    print(f'Silhouette Score for k = {k}:  {score}')
    
```

```python
plt.figure()
plt.plot(k_range, silscores_list, color='purple')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for K-means Demographic Dataset')
plt.show()
```

Based on both silhouette scores and inertia for different k-values, 4 or 5 looks like the ideal number of clusters.

#### Clustering with Ideal 'K' Value

```python
#Clustering with Scaled, clean demographic data 
kmeans = MiniBatchKMeans(n_clusters=2)
kmeans.fit(scaled_demographic_df)
cluster_labels = kmeans.predict(Xscaled_demographic)
```

```python
# using pca to reduce to 2d for visualization
pca = PCA(n_components=2)
in2d = pca.fit_transform(Xscaled_demographic)
```

```python
plt.figure()
plt.scatter(in2d[:,0],in2d[:,1],c=cluster_labels, alpha =.2)
plt.title('Clustering Demographic Features')
```

```python
np.sum(pca.explained_variance_ratio_) #30% of variance captured in first 2 PC
```

```python
# look at loadings

loadings = pca.components_

loadingsDF = pd.DataFrame(loadings, columns=list(scaled_demographic_df.columns))

loadingsDF.head()
```

```python
# saving cluster labels in dataframe, can be used to sort as well
demographicdf_clusterlabels = pd.DataFrame(Xscaled_demographic,columns=Xdemographic.columns).groupby(cluster_labels).mean()
```

```python
demographicdf_clusterlabels.T
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(Xdemographic,cluster_labels)
lr.score(Xdemographic,cluster_labels)
```

#### Finding Best 'K' Value

```python
# Look over a range of K values
k_ranges = np.arange(1, 20)

# Instantiate an empty list of inertias
inertia_list = []

# Iterate over ks
for k in k_ranges:
    
    # Instantiate
    k_means_model = KMeans(n_clusters=k)

    # Fit
    k_means_model.fit(Xscaled_quest)
    
    # Get inertia and add to list
    inertia_list.append(k_means_model.inertia_)  
```

#### Silhouette Score

```python
# Look over a range of K values
k_range = np.arange(2, 20)

# Instantiate an empty list of inertias
silscores_list = []

# Iterate over ks
for k in k_range:
    #print(f'We are currently on k= {k}')
    #kmeans
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_quest_df)
    score = silhouette_score(scaled_quest_df,kmeans.labels_)
    silscores_list.append(silhouette_score(scaled_quest_df, kmeans.labels_))
    print(f'Silhouette Score for k = {k}:  {score}')
```

```python
# install yellowbrick for silhouette score visualizer
#pip install yellowbrick
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

# Range of K values
k_range = np.arange(2, 10)  # Reduce the range for demonstration purposes

# Loop over k and visualize the silhouette scores for each k
for k in k_range:
    plt.figure()
    kmeans = KMeans(n_clusters=k, random_state=42)
    visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')
    
    # Fit the visualizer
    visualizer.fit(scaled_quest_df)  
    
    # Show the plot for each k
    visualizer.show()
```
